```python
# 1. Replace blanks with recurring values or specified values (no column here is blank)
# 2. Fix dates (no date column)
# Bonus: spelling for clarity
# 3. Remove spaces (trim, clean)
# 4. Diagnostics (istext/isnumber, len, code)
# 5. Characters (substitute, char, covert to value or text)
# 6. Duplicates

#ID
df.ID.isnull().sum()
```

```python
df2 = df.drop_duplicates()
```

```python
df2.info()
```

```python
df2
```

```python
#Marital Status
df3=df2.rename(columns={'Marital Status':'MaritalStatus','Home Owner':'HomeOwner','Communte Distance':'CommuteDistance','Purchased Bike':'PurchasedBike'})
```

```python
df3
```

```python
df3['MaritalStatus'] = df3.MaritalStatus.replace('M','Married').replace('S','Single')
```

```python
#Gender
df3['Gender'] = df3.Gender.replace('F','Female').replace('M','Male')
```

```python
#Income
df3['Income'] = df3.Income.str[1:]
```

```python
df3
```

```python
df3['AgeBracket'] = np.where(df3.Age<31, 'Adolescent', np.where(df3.Age>54,'Old','Middle Age'))
```
